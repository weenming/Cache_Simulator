```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
x = np.linspace(-3, 10, 14)
def f2(x):
    if x < 2:
        return 0.1 * x
    elif 2 <= x < 3:
        return 0.8 * (x - 2) + 0.1 * 2
    elif x >= 3:
        return 0.1 * (x - 3) + 0.1 * 2 + 0.8 * 1
def f6(x):
    if x < 6:
        return 0.1 * x
    elif 6 <= x < 7:
        return 0.8 * (x - 6) + 0.1 * 6
    elif x >= 7:
        return 0.1 * (x - 7) + 0.1 * 6 + 0.8 * 1

def fn(x, n):
    res = 0
    for ni, pre_ni in zip(n[1:], n[:-1]):
        if x > ni:
            res += 0.8
        res += (ni - pre_ni) * 0.1
    return (x - ni) * 0.1 + res 

ax.plot(x, [fn(xi, [-2, -1, 2, 4, 6]) for xi in x], ls='-', marker='x', c='steelblue', label='other strategy', zorder=1+3)
ax.plot(x, [fn(xi, [-2, -1, 6, 8, 4]) for xi in x], ls=':', marker='.', c='orange', label='Baledy algo.', zorder=2+3)
ymin, ymax = ax.get_ylim()
ax.plot([7, 7], [ymin, ymax], label='access '+'$x_{m1}$', ls='--', c='grey', alpha=0.5, zorder=0)
ax.set_xlabel('mem access count')
ax.set_ylabel('accumulated time / a.u.')
ax.legend()
fig.savefig('Baledy_demo_2.png', dpi=400)
```
